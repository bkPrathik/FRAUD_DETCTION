```python
# ══════════════════════════════════════════════════════════════════════════
# DEVICE RESOLVER  ──  called once at the top of run_fraud_tuning()
# ══════════════════════════════════════════════════════════════════════════
def _resolve_device() -> torch.device:
    """
    Auto-detects the best available device.

    Priority:
        1. FORCE_CPU=1 env var  → always CPU  (set this in your Dockerfile)
        2. CUDA available        → NVIDIA GPU
        3. MPS  available        → Apple Silicon GPU (M1/M2/M3/M4)
        4. fallback              → CPU

    Docker usage  (add to Dockerfile or docker-compose.yml):
        ENV FORCE_CPU=1

    Returns:
        torch.device
    """
    if os.environ.get("FORCE_CPU", "0") == "1":
        device = torch.device("cpu")
        print("  Device : CPU  (FORCE_CPU=1 detected — Docker / CI mode)")
        return device

    if torch.cuda.is_available():
        device = torch.device("cuda")
        print(f"  Device : CUDA  ({torch.cuda.get_device_name(0)})")
        return device

    if torch.backends.mps.is_available() and torch.backends.mps.is_built():
        device = torch.device("mps")
        print("  Device : MPS  (Apple Silicon GPU — Metal acceleration active)")
        return device

    device = torch.device("cpu")
    print("  Device : CPU  (no GPU found)")
    return device
```
