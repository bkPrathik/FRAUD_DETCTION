```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import json
import os
```

```python
data= pd.read_csv('/Users/prathikkundaragi/Downloads/fraud_dataset.csv', low_memory=True)
data.head(10)
```

```python
#removng types with no fraud cases
non_zero_fraud= data.groupby('type')['isFraud'].sum()[data.groupby('type')['isFraud'].sum()!=0]
print(non_zero_fraud.index)

data = data[data['type'].isin(non_zero_fraud.index)]
data.head(10)
```

```python
data.drop(['isFlaggedFraud','nameOrig','step','nameDest'], axis=1, inplace=True)
```

```python
data.shape
```

```python
segment = data['type'].value_counts()
segment
```

```python
# One-hot encode the 'type' column
data = pd.get_dummies(data, columns=['type'], drop_first=False, dtype=int)
data.head()
```
